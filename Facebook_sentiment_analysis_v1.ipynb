```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import io
import unicodedata
import re
import string
from numpy import linalg
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import webtext
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
df=pd.read_csv("C:/Users/DELL/Desktop/SEM7/SIN/comment_data.csv")
text=df['Comment']
print(text)
```

```python
def clean(text):
    # removing paragraph numbers
    text = re.sub('[0-9]+.\t','',str(text))
    # removing new line characters
    text = re.sub('\n ','',str(text))
    text = re.sub('\n',' ',str(text))
    # removing apostrophes
    text = re.sub("'s",'',str(text))
    # removing hyphens
    text = re.sub("-",' ',str(text))
    text = re.sub("â€” ",'',str(text))
    text = re.sub("|",'',str(text))
    # removing quotation marks
    text = re.sub('\"','',str(text))
    # removing salutation
    text = re.sub("Mr\.",'Mr',str(text))
    text = re.sub("Mrs\.",'Mrs',str(text))
    # removing any reference to outside text
    text = re.sub("[\(\[].*?[\)\]]", "", str(text))
    #removing !
    text = re.sub('!','',str(text))
    
    return text
```

```python
#tokenizing
for i in text:
    sent_tokenizer = PunktSentenceTokenizer(i)
    sents = sent_tokenizer.tokenize(i)
    print(sent_tokenize(i))
    print("\n")
    print(word_tokenize(i))
    print("\n")
    print("---------------")
```

```python
#stemming
for i in text:
    porter_stemmer = PorterStemmer()
    nltk_tokens = nltk.word_tokenize(i)
    for w in nltk_tokens:
        print (porter_stemmer.stem(w))
```

```python
#lemmatizing
for i in text:
    wordnet_lemmatizer = WordNetLemmatizer()
    nltk_tokens = nltk.word_tokenize(i)
    for w in nltk_tokens:
        print (wordnet_lemmatizer.lemmatize(w))
```
